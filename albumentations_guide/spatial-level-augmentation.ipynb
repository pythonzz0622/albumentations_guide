```python
import os
from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))
import albumentations as A
import matplotlib.pyplot as plt
import cv2 
from pycocotools.coco import COCO
import numpy as np 
A.__version__
```

```python
coco = COCO("dataset.json")
```

```python
def trans(transformer):
    transform = transformer
    augmented_image = transform(image=IMG)['image']
    label = str(transformer).split('(')[0]
    return augmented_image , label
```

```python
plt.figure(figsize = (15 , 5))
ax = plt.subplot(1 , 3 , 1)
IMG = cv2.imread('./ref_image.jpg')
IMG = IMG[: , ::-1] 
plt.title('Origin') ; plt.axis('off') ; plt.imshow(IMG) ;

ax = plt.subplot(132)
plt.imshow(IMG); plt.axis('off') ; plt.title('Mask')
annIds = coco.getAnnIds(imgIds=1, catIds=0, iscrowd=None)
anns = coco.loadAnns(annIds)
coco.showAnns(anns )

ax = plt.subplot(133)
IMG_mask = IMG.copy()
for i in range(len(coco.loadAnns(annIds))):
    temp = coco.loadAnns(annIds)[i]['bbox']
    cv2.rectangle(IMG_mask , (temp[0] ,  temp[1] ) , (temp[0] + temp[2] ,   temp[1] + temp[3]) , (255 , 255, 0) , 1)
plt.imshow(IMG_mask) ; plt.title('BBOX') ; plt.axis ; plt.show()
```

```python
annIds = coco.getAnnIds(imgIds=1, iscrowd=None)
MASK = np.zeros(IMG.shape[:-1], dtype = np.uint8)
for Ids in annIds:
    anns = coco.loadAnns(Ids)[0]
    MASK += coco.annToMask(anns)
plt.imshow(MASK)
plt.show()
```

```python
annIds = coco.getAnnIds(imgIds=1, iscrowd=None)
data = coco.loadAnns(annIds)
bbox_list = []
for i in range(5):
    xmin , ymin , width , height = data[i]['bbox']
    bbox_list.append([xmin , ymin , xmin + width , ymin + height])
```

```python
spatial_level = 'Affine,CenterCrop,CoarseDropout,Crop,CropAndPad,CropNonEmptyMaskIfExists,ElasticTransform,Flip,GridDistortion,GridDropout,HorizontalFlip,Lambda,LongestMaxSize,MaskDropout,NoOp,OpticalDistortion,PadIfNeeded,Perspective,PiecewiseAffine,PixelDropout,RandomCrop,RandomCropNearBBox,RandomGridShuffle,RandomResizedCrop,RandomRotate90,RandomScale,RandomSizedBBoxSafeCrop,RandomSizedCrop,Resize,Rotate,SafeRotate,ShiftScaleRotate,SmallestMaxSize,Transpose,VerticalFlip'
spatial_level = spatial_level.split(',')
other_level =  ['CenterCrop', 'Crop' , 'CropAndPad' , 'CropNonEmptyMaskIfExists' , 'MaskDropout' , 'RandomCrop' , 'RandomCropNearBBox' , 'RandomResizedCrop' , 'RandomSizedBBoxSafeCrop' , 'RandomSizedCrop' , 'Resize']
spatial_level = list(set(spatial_level) - set(other_level))
```

```python
aug_1 = [
A.CenterCrop(height = 100 , width  = 100 , always_apply= True ),
A.Crop(x_min=0, y_min=0, x_max=100, y_max=100, always_apply= True),
A.CropAndPad(px = 10 , always_apply= True),
A.RandomResizedCrop(height = 100  , width = 100 , always_apply = True),
A.RandomSizedCrop(min_max_height= (20 , 100) , height = 100 , width = 100 , always_apply = True)
]

aug_mask = [
A.CropNonEmptyMaskIfExists(height = 100 , width = 100 , always_apply = True ) ,
A.MaskDropout(always_apply = True)
]

aug_bbox = [
A.RandomSizedBBoxSafeCrop(height = 100 , width = 100 , always_apply = True),
A.RandomCropNearBBox(always_apply = True),
A.RandomSizedBBoxSafeCrop(height = 100 , width = 100 , always_apply = True)
]
```

```python
fig , ax = plt.subplots(5, 5, figsize = ( 15 , 15))
c = 0
for i in range(5):
    for j in range(5):
        try:
            transformer = getattr(A , spatial_level[c])(always_apply = True)
            img , label = trans(transformer)
            ax[i, j].set_title(label)
            ax[i, j ].axis('off')
            ax[i, j ].imshow(img)
            c +=1
        except IndexError:
            break
plt.show()

fig , ax = plt.subplots(1, 5, figsize = ( 15 , 3))
c = 0
for i in range( 5):
    transformer = aug_1[i]
    img , label = trans(transformer)
    ax[i].set_title(label)
    ax[i].axis('off')
    ax[i].imshow(img)
plt.show()
```

```python
for i in range(2):
    aug =  aug_mask[i](image = IMG , mask = MASK)
    fig , ax = plt.subplots(1, 4, figsize = ( 12 , 3))
    fig.suptitle(str(aug_mask[i]).split('(')[0])
    ax[0].imshow(IMG)
    ax[1].imshow(MASK)
    ax[2].imshow(aug['image'])
    ax[3].imshow(aug['mask'])
    plt.show()
```

```python
aug_bbox = [
A.RandomSizedBBoxSafeCrop(height = 100 , width = 100 , always_apply = True),
A.RandomSizedBBoxSafeCrop(height = 100 , width = 100 , always_apply = True)
]
```

```python
for i in range(2):
    aug =  aug_bbox[i](image = IMG , bboxes = bbox_list )
    fig , ax = plt.subplots(1, 4, figsize = ( 12 , 3))
    fig.suptitle(str(aug_bbox[i]).split('(')[0])
    ax[0].imshow(IMG)
    IMG_BBOX = IMG.copy()
    for temp in bbox_list:
        cv2.rectangle(IMG_BBOX , (int(temp[0]) ,  int(temp[1]) ) , (int(temp[2]) , int(temp[3])) , 
                      (255 , 255, 0) , 1)
    ax[1].imshow(IMG_BBOX)
    ax[2].imshow(aug['image'])
    for temp in aug['bboxes']:
        cv2.rectangle(IMG_BBOX , (int(temp[0]) ,  int(temp[1]) ) , (int(temp[2]) , int(temp[3])) , 
                      (255 , 255, 0) , 1)
    ax[3].imshow(IMG_BBOX)
    plt.show()
```
